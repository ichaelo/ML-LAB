```python
# Genius (aggregation without pandas)
import csv
```

```python

with open("/content/dataset/global-data-on-sustainable-energy.csv") as f:
    reader = csv.reader(f)
    headers = next(reader) #first string
    aggregate_data = {}

    for row in reader: # run for each string of csv-file
        # ÐŸÑ€Ð¾Ð²ÐµÑ€ÑÐµÐ¼, ÐµÑÐ»Ð¸ ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ð¹ Ð² ÑÑ‚Ñ€Ð¾ÐºÐµ ÑÐ¾Ð²Ð¿Ð°Ð´Ð°ÐµÑ‚ Ñ ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾Ð¼ ÐºÐ¾Ð»Ð¾Ð½Ð¾Ðº
        if len(row) == len(headers): # if num znach == quantity kolonok
            for i in range(len(row)): # run for each znach from string
                # number or categorical
                try:
                    # in number
                    value = float(row[i])

                    # if znach - number sign, find categorical sign
                    category = headers[i]

                    # category in slovar ? add number sign
                    if category in aggregate_data:
                        aggregate_data[category].append(value)
                    # no category - create new with number sign
                    else:
                        aggregate_data[category] = [value]
                except ValueError:
                    # znach no number - skip
                    continue
        else:
            continue

# Ð’Ñ‹Ñ‡Ð¸ÑÐ»ÑÐµÐ¼ Ð¼ÐµÐ´Ð¸Ð°Ð½Ñƒ Ð¸ ÑÑ€ÐµÐ´Ð½ÐµÐµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ Ð´Ð»Ñ ÐºÐ°Ð¶Ð´Ð¾Ð¹ ÐºÐ°Ñ‚ÐµÐ³Ð¾Ñ€Ð¸Ð¸
for category, values in aggregate_data.items(): #  items() -- get pair key - meaning
    median = sorted(values)[len(values) // 2]
    mean = sum(values) / len(values)
    print(f'Category: {category}')
    print(f'Median: {median}')
    print(f'Mean: {mean}')
    print('---')

```
