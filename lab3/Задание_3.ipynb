##Поиск выбросов и аномалий

**Те, кто выполняет задания со * делают поиск выбросов с помощью написанного ранее класса**

Найти выбросы по числовой переменной (использовать метод основанный на среднем и на межквартильном размахе). Объясните результат

Вывести статистику с выбросами и без. Сравните и объясните результат


```python
from statistics import median, mean
import math
import numpy as np
class Outline:
  def __init__(self, df):
    self.df = df

  def delete_outliers_by_mean_and_std(self, column_name):
    median = self.df[column_name].median()
    mean = self.df[column_name].mean()
    std = self.df[column_name].std()
    quantity = self.df[column_name].count()

    if median - mean <= 3 * math.sqrt(std**2/quantity):
      coff = 3
    else:
      coff = 5

    interval_left = mean - coff*std
    interval_right = mean + coff*std
    outlier_indices = []

    for k, v in self.df[column_name].items():
      if v < interval_left and v > interval_right:
        outlier_indices.append(v)
    return outlier_indices


  def delete_outliers_by_qvartil(self, column_name):
    quan25 = self.df[column_name].quantile(0.25)
    quan75 = self.df[column_name].quantile(0.75)
    quan_diff = quan75 - quan25
    left_border = quan25 - 1.5 * quan_diff
    right_border = quan75 + 1.5 * quan_diff

    outlier_indices = []

    for k, v in self.df[column_name].items():
      if v < left_border and v > right_border:
        outlier_indices.append(v)

    return outlier_indices
```

```python
#Методом среднего
outlier1 = Outline(df_drop)
del_by_mean = outlier1.delete_outliers_by_mean_and_std('Renewable energy share in the total final energy consumption (%)')
del_by_mean
```

```python
#При помощи межквартильного размаха
outlier2 = Outline(df_drop)
del_by_qvartil = outlier2.delete_outliers_by_qvartil('Renewable energy share in the total final energy consumption (%)')
del_by_qvartil
```

Реализовать расчет корреляции Спирмана

```python
# Корреляция Спирмена

def spearman_rank_correlation(df, column1, column2):
    # Получаем ранги для каждого столбца
    rank1 = df[column1].rank()
    rank2 = df[column2].rank()

    # Рассчитываем разницу в рангах и ее квадрат
    rank_diff = rank1 - rank2

    # Вычисляем сумму квадратов разностей в рангах (D)
    sum_squared_diff = (rank_diff ** 2).sum()

    # Вычисляем число наблюдений
    n = len(df)

    # корреляция Спирмена
    correlation = 1 - (6 * sum_squared_diff) / (n * (n ** 2 - 1))

    # Создаем новый DataFrame с результатами
    result_df = pd.DataFrame({
        column1: [1,correlation],
        column2: [correlation,1]
    }, index=[column1, column2])

    return result_df
```

```python
df = pd.read_csv('/content/dataset/global-data-on-sustainable-energy.csv')
'''data = {'A': [1, 2, 3, 4, 5],
        'B': [1, 3, 2, 4, 5],
        'C': [5, 4, 3, 2, 1]}
df = pd.DataFrame(data)'''
correlation_matrix = spearman_rank_correlation(df, 'Renewable energy share in the total final energy consumption (%)', 'Renewable-electricity-generating-capacity-per-capita')
correlation_matrix
```

```python
import pandas as pd
import numpy as np


def corr_pearson(column_1, column_2):
  cov = np.cov(column_1, column_2)
  return cov/(column_1.std() * column_2.std())
```
